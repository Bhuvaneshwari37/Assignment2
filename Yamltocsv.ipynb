```python
import yaml
import glob
import os
import pandas as pd

# Step 1: Get the list of all YAML files
yaml_files = glob.glob("D:/vscode/project2/project 2/**/*.yaml", recursive=True)

def yaml_load(yaml_path):
    """Load a YAML file and return its data."""
    with open(yaml_path, 'r') as file:
        yaml_data = yaml.safe_load(file)
    return yaml_data

# Step 2: Load data from each YAML file and process dates
all_data = []

for file in yaml_files:
    data = yaml_load(file)
    
    # Extract date and symbol (Ticker) from YAML file
    if 'date' in data and 'Ticker' in data:
        try:
            data['date'] = pd.to_datetime(data['date'])  # Convert to pandas datetime
            data['file_name'] = os.path.basename(file)  # Add filename for reference
            all_data.append(data)
        except Exception as e:
            print(f"Error parsing date in {file}: {e}")

# Step 3: Convert data into a DataFrame
df = pd.DataFrame(all_data)

if not df.empty:
    df['month'] = df['date'].dt.strftime('%Y-%m')  # Extract month as "YYYY-MM"
    
    # Create a new column 'symbol' (from Ticker) for extracting symbols
    df['symbol'] = df['Ticker'].apply(lambda x: x)  # This step can be customized as per your symbol extraction logic

    # Step 4: Save data grouped by month and symbol
    output_dir = "D:/vscode/project2/output_csv"
    os.makedirs(output_dir, exist_ok=True)  # Ensure output directory exists

    for (month, symbol), group in df.groupby(["month", "symbol"]):
        csv_path = os.path.join(output_dir, f"{month}_{symbol}.csv")
        group.drop(columns=["month", "symbol"]).to_csv(csv_path, index=False)  # Save each month-symbol group as a separate CSV
        print(f"Saved {csv_path}")
else:
    print("No valid YAML data found.")
```

```python
import yaml
import glob
import os
import pandas as pd

# Step 1: Get the list of all YAML files
yaml_files = glob.glob("D:/vscode/project2/project 2/**/*.yaml")

def yaml_load(yaml_path):
    with open(yaml_path, 'r') as file:
        yaml_data = yaml.safe_load(file)
    return yaml_data

# Specify the months and initialize a DataFrame
months = ["2023-10", "2023-11"]
path_date = []

for month in months:
    path_date.append(pd.date_range(start=f'{month}-01', end=f'{month}-28'))  # Adjust the end date as per the month

# Step 2: Load data from each YAML file and process it
all_data = []

for file in yaml_files:
    data = yaml_load(file)
    # Check if the 'date' in YAML file matches the specified months
    date = pd.to_datetime(data['date'])
    if any(date in pd.Series(dates) for dates in path_date):
        all_data.append(data)

# Step 3: Convert data into a DataFrame
df = pd.DataFrame(all_data)

if not df.empty:
    # Extract month from the date
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.strftime('%Y-%m')

    # Step 4: Save data grouped by month and ticker (symbol)
    output_dir = "D:/vscode/project2/output_csv"
    os.makedirs(output_dir, exist_ok=True)  # Ensure output directory exists

    for (month, ticker), group in df.groupby(["month", "Ticker"]):
        csv_path = os.path.join(output_dir, f"{month}_{ticker}.csv")
        group.drop(columns=["month"]).to_csv(csv_path, index=False)  # Save each month-ticker group as a separate CSV
        print(f"Saved {csv_path}")
else:
    print("No valid YAML data found.")
```
